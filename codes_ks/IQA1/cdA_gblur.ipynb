```python
import keras
from keras.datasets import cifar10
(x_train, _), (x_test, _) = cifar10.load_data()
```

```python
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
x_train = np.reshape(x_train, (len(x_train), 32, 32, 3))  # adapt this if using `channels_first` image data format
x_test = np.reshape(x_test, (len(x_test), 32, 32, 3))  # adapt this if using `channels_first` image data format
```

```python
# Adding GBLUR
x_train, x_train_noisy = utils.cifar10_gblur(x_train)
x_test, x_test_noisy = utils.cifar10_gblur(x_test)
```

```python
# Building the graph
L = 5
Fc = [3, 32, 64, 128, 256, 512]     # Number of clean filters including input
Fn = [3, 16, 32, 64, 128, 256]      # Number of distorted filters including input

g = tf.Graph()

with g.as_default():
    # Input placeholders
    clean_input = tf.placeholder(tf.float32, shape = [None, 32, 32, 3])
    dist_input = tf.placeholder(tf.float32, shape = [None, 32, 32, 3])

    clean_encs = []
    clean_weights = []
    clean_enc = clean_input
    for i in range(L):
    # Weights
        W_clean_init = tf.random_uniform([3, 3, Fc[i], Fc[i+1]], 
                                         -1.0 / math.sqrt(Fc[i]), 
                                         1.0 / math.sqrt(Fc[i+1]))
        W_clean = tf.Variable(W_clean_init)
        clean_weights.append(W_clean)
    # Layers
        clean_enc = tf.nn.conv2d(clean_enc, W_clean, strides=[1, 2, 2, 1], padding='SAME')
        clean_enc = tf.nn.relu(clean_enc)
        clean_encs.append(clean_enc)

    clean_dec = clean_encs[L-1]
    for i in range(L-1)[::-1]:
    # Weights
        W_clean = clean_weights[i+1]
    # Layers
        clean_dec = tf.nn.conv2d_transpose(clean_dec, W_clean, tf.shape(clean_encs[i]), 
                                          strides=[1, 2, 2, 1], padding='SAME')
        clean_dec = tf.nn.relu(clean_dec)
        clean_dec = tf.add(clean_dec, clean_encs[i])
```
