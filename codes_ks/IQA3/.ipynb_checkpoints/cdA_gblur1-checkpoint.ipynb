```python
from keras.datasets import cifar10
(x_train, _), (x_test, _) = cifar10.load_data()
```

```python
x_train.shape, x_test.shape
```

```python
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
x_train = np.reshape(x_train, (len(x_train), 32, 32, 3))  # adapt this if using `channels_first` image data format
x_test = np.reshape(x_test, (len(x_test), 32, 32, 3))  # adapt this if using `channels_first` image data format
```

```python
# Adding GBLUR
x_train, x_train_noisy = utils.cifar10_gblur(x_train)
x_test, x_test_noisy = utils.cifar10_gblur(x_test)
```

```python
x_train.shape, x_train_noisy.shape, x_test.shape, x_test_noisy.shape
```

```python
# Displaying noisy images
n = 10
plt.figure(figsize=(20, 4))
for i in range(1, n):
    # display original
    ax = plt.subplot(2, n, i)
    plt.imshow(x_train[i].reshape(32, 32, 3))
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # display reconstruction
    ax = plt.subplot(2, n, i + n)
    plt.imshow(x_train_noisy[i].reshape(32, 32, 3))
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
```

```python
import keras
from keras import losses
from keras.models import Model, load_model
from keras.layers import Input, Dense, Conv2D, MaxPooling2D, UpSampling2D, concatenate

L = 5
Fc = [64, 128, 128, 256, 512] # Number of filters at each layer of clean encoder
Fn = [16, 32, 64, 64, 128]    # Number of filters at each layer of noise encoder
# Clean Encoder
clean_enc_inp = Input(shape = (None, None, 3))
clean_h = Conv2D(Fc[0], (3, 3), activation='relu', padding='same', strides=(2, 2))(clean_enc_inp)
for convs in range(1, L):
    filters = Fc[convs]
    clean_h = Conv2D(filters, (3, 3), activation='relu', padding='same', strides=(2, 2))(clean_h)
clean_encoder = Model(inputs=[clean_enc_inp], outputs=[clean_h])
clean_encoder.compile(optimizer='adadelta', loss='mean_squared_error')

# noise Encoder
noise_enc_inp = Input(shape = (None, None, 3))
noise_h = Conv2D(Fn[0], (3, 3), activation='relu', padding='same', strides=(2, 2))(noise_enc_inp)
for convs in range(1, L):
    filters = Fn[convs]
    noise_h = Conv2D(filters, (3, 3), activation='relu', padding='same', strides=(2, 2))(noise_h)
noise_encoder = Model(inputs=[noise_enc_inp], outputs=[noise_h])
noise_encoder.compile(optimizer='adadelta', loss='mean_squared_error')

# Decoder
h = Input(shape = (None, None, Fc[L-1] + Fn[L-1]))
dec = UpSampling2D((2, 2))(h)
dec = Conv2D(Fc[L-1] + Fn[L-1], (3, 3), activation='relu', padding='same')(dec)
for deconvs in range(1, L):
    if deconvs == L-1:
        filters = 3
    else:
        filters = Fc[::-1][deconvs] + Fn[::-1][deconvs]
    dec = UpSampling2D((2, 2))(dec)
    dec = Conv2D(filters, (3, 3), activation='relu', padding='same')(dec)
decoder = Model(inputs=[h], outputs=[dec])
decoder.compile(optimizer='adadelta', loss='mean_squared_error')

# Fitting clean images
clean_input = Input(shape = (None, None, 3))
clean_noise = Input(shape = (None, None, Fn[L-1]))
clean = clean_encoder(clean_input)
clean_enc = concatenate([clean, clean_noise], axis=-1)

# Fitting distorted images
dist_input = Input(shape = (None, None, 3))
dist_noise = noise_encoder(dist_input)
dist_enc = concatenate([clean, dist_noise], axis=-1)

clean_pred = decoder(clean_enc)
dist_pred = decoder(dist_enc)

IQA_model = Model(inputs=[clean_input, clean_noise, dist_input], outputs=[clean_pred, dist_pred])
IQA_model.compile(optimizer='adadelta', loss='mean_squared_error')
```
